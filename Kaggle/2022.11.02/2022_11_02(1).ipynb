```python
# T1-10. Yeo-Johnson and Box–Cox / #여존슨 #박스-콕스 #결측치 #최빈값
# 주어진 데이터에서 20세 이상인 데이터를 추출하고 
# 'f1'컬럼을 결측치를 최빈값으로 채운 후, 
# f1 컬럼의 여-존슨과 박스콕스 변환 값을 구하고, 
# 두 값의 차이를 절대값으로 구한다음 
# 모두 더해 소수점 둘째 자리까지 출력(반올림)하시오
# 풀이

import pandas as pd
import numpy as np
from sklearn.preprocessing import 여-존슨, 박스-콕스

df = pd.read_csv('./data/basic1.csv')
# display(df.head(3))
# print(df.info())
# print(df.isnull().sum())
# print(df.shape)

df = df[df['age'] >= 20]
# display(df)

df['f1'] = df['f1'].fillna(df['f1'].최빈값())
# dir(sklearn.preprocessing)
# dir(df)

box_cox = 박스-콕스()
yeo_johnson = 여-존슨()
b = df['f1'].box_cox()
y = df['f1'].yeo_johnson()

answer = abs(b - y)
answer = round(answer, 2)
print(answer)
```

```python
# 해답

import pandas as pd
import numpy as np
from sklearn.preprocessing import power_transform

df = pd.read_csv('./data/basic1.csv')

df = df[df['age'] >= 20]

# df['f1'].mode()
mode_ = df['f1'].mode()[0]
df['f1'] = df['f1'].fillna(mode_)
# df.isnull().sum()

# method 디폴트 값은 여-존슨’yeo-johnson’
# df['yeo'] = power_transform(df[['f1']])

# standardize=False는 python과 R의 디폴트가 달라 결과를 맞추기 위함이었어요! 
# 이번 문제는 출제 가능성이 낮아 보여요 (이미 풀이한 분들이 있어 그대로 두었습니다.)
df['yeo'] = power_transform(df[['f1']], standardize=False)

# df['yeo'].head()

# df['box'] = power_transform(df[['f1']], method='box-cox')
df['box'] = power_transform(df[['f1']], method='box-cox', standardize=False)
# df['box'].head()

answer = round(sum(np.abs(df['yeo'] - df['box'])), 2)
print(answer)
```

```python
# T1-11. min-max scaling / #스케일링 #상하위값
# min-max스케일링 기준 상하위 5% 구하기
# 주어진 데이터에서 'f5'컬럼을 min-max 스케일 변환한 후,
# 상위 5%와 하위 5% 값의 합을 구하시오
# 풀이

import pandas as pd
import numpy as np
from sklearn.preprocessing import minmax_scale

df = pd.read_csv('./data/basic1.csv')
display(df.head(3))
# print(df.info())
# print(df.isnull().sum())
# print(df.shape)

df['f5'] = minmax_scale(df['f5'])
df = df.sort_values('f5', ascending=False)
# df

sum1 = sum(df.iloc[:int(len(df)*0.05),:]['f5'])
sum2 = sum(df.iloc[int(len(df)*0.95):,:]['f5'])

answer = sum1 + sum2
print(answer)
```

```python
# T1-12. top10-bottom10 / #그룹핑 #정렬 #상하위값
# 주어진 데이터에서 상위 10개 국가의 접종률 평균과 하위 10개 국가의 접종률 평균을 구하고, 
# 그 차이를 구해보세요 (단, 100%가 넘는 접종률 제거, 소수 첫째자리까지 출력)
# 풀이

import pandas as pd
import numpy as np

df = pd.read_csv('./data/covid-vaccination-vs-death_ratio.csv', index_col=0)
display(df.head(3))
print(df.info())
print(df.isnull().sum())
print(df.shape)
# help(pd.read_csv)

# df['country'].unique()
df = df.sort_values('ratio', ascending=False)
index_ = df['ratio'][df['ratio'] > 100].index
df = df.drop(index_)

upper = df['ratio'][:10].mean()
lower = df['ratio'][-10:].mean()
# df['ratio'][-10:]

answer = round(upper - lower, 1)
print(answer)
```

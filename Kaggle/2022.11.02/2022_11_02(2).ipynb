```python
# 해답

import pandas as pd
import numpy as np

df = pd.read_csv('./data/winequality-red.csv')

df_corr = df.corr()
df_corr = df_corr[:-1]
display(df_corr['quality'])

# 상관관계가 가장 큰 값과 가장 작은 값 (절대값으로 확인)
# 이거 왜 이렇게 하지? 가장 큰값과 가장 작은 값을 더하라매 절댓값 씌우면 안되는거 아님?
max_corr = abs(df.corr()['quality'][:-1]).max()
min_corr = abs(df.corr()['quality'][:-1]).min()

if max_corr not in df.corr()[['quality']][:-1].values:
    max_corr=-max_corr
if min_corr not in df.corr()[['quality']][:-1].values:
    min_corr=-min_corr
    
# 결과 출력
ans=round(max_corr+min_corr,2)
print(ans)

# 아 상관관계가 가장 큰 값과 가장 작은 값이네 (+)든 (-)든 절댓값으로 보란 얘기였구나
```

```python
# 풀이 수정

import pandas as pd
import numpy as np

df = pd.read_csv('./data/winequality-red.csv')
display(df.head(3))
# print(df.info())
# print(df.isnull().sum())
# print(df.shape)

df2 = df.corr()['quality'].sort_values(ascending=False)
df2 = df2[1:]
display(df2)

# answer = df2[1] + df2[len(df2)-1]
# answer = round(answer, 2)
# print(answer)

display(abs(df2).sort_values(ascending=False))
max_corr_idx = abs(df2).sort_values(ascending=False).index[0]
min_corr_idx = abs(df2).sort_values(ascending=False).index[-1]

max_corr = df2[max_corr_idx]
min_corr = df2[min_corr_idx]

answer = max_corr + min_corr
answer = round(answer, 2)
print(answer)
```

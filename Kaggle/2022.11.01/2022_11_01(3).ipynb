```python
# T1-8. Cumulative Sum / #누적합 #결측치 #보간
# 주어진 데이터 셋에서 'f2' 컬럼이 1인 조건에 해당하는 데이터의 'f1'컬럼 누적합을 계산한다. 
# 이때 발생하는 누적합 결측치는 바로 뒤의 값을 채우고, 누적합의 평균값을 출력한다. 
# (단, 결측치 바로 뒤의 값이 없으면 다음에 나오는 값을 채워넣는다)
# 풀이

import pandas as pd
import numpy as np

df = pd.read_csv('./data/basic1.csv')
display(df.head(3))
print(df.info())
# print(df.isnull().sum())
# print(df.shape)

print(help(df.fillna))

answer = df[df['f2'] == 1]['f1'].sum()
print(answer)
```

```python
# 해답

import pandas as pd
import numpy as np

df = pd.read_csv('./data/basic1.csv')

df2 = df[df['f2'] == 1]['f1'].cumsum()
display(df2)

df2 = df2.fillna(method='bfill')
display(df2)

answer = df2.mean()
print(answer)
```

```python
# 풀이 수정

import pandas as pd
import numpy as np

df = pd.read_csv('./data/basic1.csv')
display(df.head(3))
print(df.info())
# print(df.isnull().sum())
# print(df.shape)

# cumsum() : 누적합
# cumprod() : 누적곱
df2 = df[df['f2']==1]['f1'].cumsum()
display(df2)

# method = bfill (결측값을 바로 아래 값과 동일하게 변경)
# method = ffill (결측값을 바로 위 값과 동일하게 변경)
df2 = df2.fillna(method='bfill')

answer = df2.mean()
print(answer)
```

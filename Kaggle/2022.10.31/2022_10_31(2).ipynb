```python
# 풀이 수정

import pandas as pd
import numpy as np

df = pd.read_csv('./data/basic1.csv')

display(df.head(3))
print(df.info())
print(df.isnull().sum())
print(df.shape)
# print(df.columns)

# len(df)*0.8
drop_cols = list(df.isnull().sum()[df.isnull().sum() >= len(df)*0.8].index)
print(drop_cols)

df = df.drop(columns=drop_cols)
display(df.head(3))

# df.groupby('city').median()
k = df[df['city'] == '경기']['f1'].median()
d = df[df['city'] == '대구']['f1'].median()
b = df[df['city'] == '부산']['f1'].median()
s = df[df['city'] == '서울']['f1'].median()
# s, k, b, d

df['f1'] = df['f1'].fillna(df['city'].map({'경기': k, '대구': d, '부산': b, '서울': s}))

answer = df['f1'].mean()
print(answer)
```

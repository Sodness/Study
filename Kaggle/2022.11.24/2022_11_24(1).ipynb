```python
# T1-18. Time-Series2 / #주말 #평일 #비교 #시계열데이터
# 주어진 데이터에서 2022년 5월 주말과 평일의 sales컬럼 평균값 차이를 구하시오 
# (소수점 둘째자리까지 출력, 반올림)

import pandas as pd

df = pd.read_csv('./data/basic2.csv')
# print(df.head(3)) # Date(object)
# print(df.isnull().sum()) # UV(46)
# print(df.shape) # (730, 5)
# print(df.info())

df['Date'] = pd.to_datetime(df['Date'])
df['year'] = df['Date'].dt.year
df['month'] = df['Date'].dt.month
df['dayofweek'] = df['Date'].dt.dayofweek # 0(Monday) ~ 6(Sunday)
# print(df.head())
# print(help(df['Date'].dt.dayofweek))

cond1 = df['year'] == 2022
cond2 = df['month'] == 5
df = df[cond1 & cond2]
# print(df.head())

cond1 = df['dayofweek'] == 5
cond2 = df['dayofweek'] == 6
mean1 = df[cond1 | cond2]['Sales'].mean()
# print(df[~(cond1 | cond2)])
mean2 = df[~(cond1 | cond2)]['Sales'].mean()

answer = round(mean1 - mean2, 2)
print(answer)
```

```python
# T1-19. Time-Series3 (monthly total) / #월별 #총계 #비교 #데이터값변경
# 주어진 데이터에서 2022년 월별 Sales 합계 중 가장 큰 금액과
# 2023년 월별 Sales 합계 중 가장 큰 금액의 차이를 절대값으로 구하시오.
# 단 Events컬럼이 '1'인경우 80%의 Salse값만 반영함
# (최종값은 소수점 반올림 후 정수 출력)
# 풀이

import pandas as pd

df = pd.read_csv('./data/basic2.csv')
# print(df.head(3))
# print(df.isnull().sum()) # UV(46)
# print(df.shape) # (730, 5)
# print(df.info()) # Date(object)

df['Date'] = pd.to_datetime(df['Date'])
df['year'] = df['Date'].dt.year
df['month'] = df['Date'].dt.month
# print(df.head(3))

df['Sales_E'] = df['Sales']
print(df[df['Events'] == 1]['Sales_E'] * 0.8)
df[df['Events'] == 1]['Sales_E'] = df[df['Events'] == 1]['Sales_E'] * 0.8
# print(df.isnull().sum())
# print(df.head(10)) # 적용이 안되네
# 1. for 돌려서 행번호 찾아서 바꾸거나
# 2. lambda : apply로 하거나
# 3. events가 1인 df.index -> ls, df[ls] = 곱하기 0.8 으로 약간 잡기술로도 가능

df_2022 = df[df['year'] == 2022]
df_2023 = df[df['year'] == 2023]

sum1 = df_2022.groupby('month').sum()['Sales_E'].max()
sum2 = df_2023.groupby('month').sum()['Sales_E'].max()

answer = round(abs(sum1 - sum2), 0)
print(answer)
```

```python
# 해답

import pandas as pd

df = pd.read_csv('./data/basic2.csv')

df['Date'] = pd.to_datetime(df['Date'])
df['year'] = df['Date'].dt.year
df['month'] = df['Date'].dt.month
df['day'] = df['Date'].dt.day
df['dayofweek'] = df['Date'].dt.dayofweek

def event_sales(x):
    if x['Events'] == 1:
        x['Sales2'] = x['Sales']*0.8
    else:
        x['Sales2'] = x['Sales']
    return x


df = df.apply(lambda x: event_sales(x), axis=1) #1일 경우 row, 0일 경우 컬럼

cond = df['year'] == 2022
df1 = df[cond]
sale1 = df1.groupby('month')['Sales2'].sum().max()

cond = df['year'] == 2023
df2 = df[cond]
sale2 = df2.groupby('month')['Sales2'].sum().max()

print(int(round(abs(sale1 - sale2),0)))
```

```python
# 풀이 수정

import pandas as pd

df = pd.read_csv('./data/basic2.csv')
# print(df.head(3))
# print(df.isnull().sum()) # UV(46)
# print(df.shape) # (730, 5)
# print(df.info()) # Date(object)

df['Date'] = pd.to_datetime(df['Date'])
df['year'] = df['Date'].dt.year
df['month'] = df['Date'].dt.month
# print(df.head(3))

def sales_E(x):
    if x['Events'] == 1:
        x['Sales_E'] = x['Sales'] * 0.8
    else:
        x['Sales_E'] = x['Sales']
    return x

df = df.apply(lambda x: sales_E(x), axis=1)

df_2022 = df[df['year'] == 2022]
df_2023 = df[df['year'] == 2023]

sum1 = df_2022.groupby('month').sum()['Sales_E'].max()
sum2 = df_2023.groupby('month').sum()['Sales_E'].max()

answer = round(abs(sum1 - sum2), 0)
print(int(answer))
```

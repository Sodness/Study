```python
import pandas as pd

X_train = pd.read_csv('./data/t2-2-X_train.csv')
y_train = pd.read_csv('./data/t2-2-y_train.csv')
X_test = pd.read_csv('./data/t2-2-X_test.csv')

display(X_train.head(3))
display(y_train.head(3))
display(X_test.head(3))
```

```python
X_train.info()
```

```python
print(X_train.isnull().sum())
print(X_test.isnull().sum())
```

```python
X_train.describe()
```

```python
X_train.corr()
```

```python
train = X_train.join(y_train.set_index('Serial No.'), how='left', on='Serial No.')
train.head(3)
```

```python
train[train.columns[1:]].corr()['Chance of Admit '].sort_values(ascending=False)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
from sklearn.model_selection import cross_val_score

# cross_val_scroe : 사이킷런 교차검증 함수
# estimator : 분류(Classifier), 회귀(Regressor) 구분 인자
# X : 훈련 데이터 세트
# Y : 훈련 데이터 레이블 세트
# scoring : 예측 성능 평가 지표 (r2 : R-squared 제곱합)
# cv : 교차 검증 폴드 수

scores = cross_val_score(lr, X_train.iloc[:,1:], y_train['Chance of Admit '], scoring='r2', cv=5)
print(scores)
print(scores.mean())
```

#### r2

```python
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(random_state=2022)

scores = cross_val_score(rf, X_train.iloc[:,1:], y_train['Chance of Admit '], scoring='r2', cv=5)
print(scores)
print(scores.mean())
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X_train.iloc[:,1:], y_train['Chance of Admit '])
pred = model.predict(X_test.iloc[:,1:])
pred
```

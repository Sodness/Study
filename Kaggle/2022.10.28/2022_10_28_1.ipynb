```python
# 문제3 - 풀이
# 데이터셋(basic1.csv)의 'age'컬럼의 이상치를 더하시오!
# 단, 평균으로부터 '표준편차*1.5'를 벗어나는 영역을 이상치라고 판단함

import pandas as pd

df = pd.read_csv('./data/basic1.csv')
# display(df.head(3))

age_mean = df['age'].mean()
age_std = df['age'].std()

sum1 = df['age'][df['age'] < (age_mean - age_std*1.5)].sum()
sum2 = df['age'][df['age'] > (age_mean + age_std*1.5)].sum()
answer = sum1 + sum2

# print((age_mean - age_std*1.5), (age_mean + age_std*1.5))
print(answer)
```

```python
# 문제3 - 해답
# 데이터셋(basic1.csv)의 'age'컬럼의 이상치를 더하시오!
# 단, 평균으로부터 '표준편차*1.5'를 벗어나는 영역을 이상치라고 판단함

import pandas as pd
import numpy as np

df = pd.read_csv('./data/basic1.csv')

std = np.std(df['age']) * 1.5
mean = df['age'].mean()

min_out = mean - std
max_out = mean + std
print(min_out, max_out)

answer = df[(df['age']>max_out) | (df['age']<min_out)]['age'].sum()
print(answer)
```

```python
# 해답2

import pandas as pd

data = pd.read_csv('./data/mtcars.csv')
display(data.head(3))

def minmax(data):
    data = (data - min(data)) / (max(data) - min(data))
    return data

data['qsec'] = minmax(data['qsec'])
display(data.head(3))

print(sum(data['qsec'] > 0.5))
```
